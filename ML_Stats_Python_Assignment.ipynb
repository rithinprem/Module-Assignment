```python
'''
1.7. Write a program to create a simple calculator performing only four basic operation
(+,-,/,*)'''

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    else:
        return x / y

def calculator():
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")

    while True:
        choice = input("Enter choice (1/2/3/4): ")

        if choice in ('1', '2', '3', '4'):
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))

            if choice == '1':
                print(f"{num1} + {num2} = {add(num1, num2)}")

            elif choice == '2':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")

            elif choice == '3':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")

            elif choice == '4':
                print(f"{num1} / {num2} = {divide(num1, num2)}")

            # Check if the user wants another calculation
            next_calculation = input("Do you want to perform another calculation? (yes/no): ")
            if next_calculation.lower() != 'yes':
                break
        else:
            print("Invalid input")

calculator()
```

```python
'''
1.9. Write a program to find the factors of a whole number using a while loop.'''

def find_factors(number):
    factors = []
    i = 1
    while i <= number:
        if number % i == 0:
            factors.append(i)
        i += 1
    return factors

# Example usage
num = int(input("Enter a whole number: "))
factors = find_factors(num)
print(f"The factors of {num} are: {factors}")
```

```python
'''
1.11. Write a program to find prime numbers between 2 to 100 using nested for loops.'''

def find_primes(start, end):
    prime_numbers = []
    for num in range(start, end + 1):
        if num > 1:  # Primes are greater than 1
            is_prime = True
            for i in range(2, int(num**0.5) + 1):
                if num % i == 0:
                    is_prime = False
                    break
            if is_prime:
                prime_numbers.append(num)
    return prime_numbers

# Find prime numbers between 2 and 100
start = 2
end = 100
primes = find_primes(start, end)
print(primes)
```

```python
'''
1.12. Write the programs for the following:
1) Accept the marks of the student in five major subjects and display the same.
2) Calculate the sum of the marks of all subjects. Divide the total marks by number of subjects(ie 5),
calculate percentage = total marks/5 and display the percentage.

3) Find the grade of the student as per the following criteria. Hint: Use Match and case for this:

Criteria                                  Grade
percentage>85                               A
percentage<85 && percentage>=75             B
percentage<75 && percentage>=50             C
percentage>30 && percentage<=50             D
percentage<30                               Reappear          '''

# 1. Accept the marks of the student in five major subjects and display the same
def get_student_marks():
    subjects = ["Subject 1", "Subject 2", "Subject 3", "Subject 4", "Subject 5"]
    marks = []
    for subject in subjects:
        mark = float(input(f"Enter marks for {subject}: "))
        marks.append(mark)
    return marks

def display_marks(marks):
    for i, mark in enumerate(marks, start=1):
        print(f"Marks for Subject {i}: {mark}")

# 2. Calculate the sum of the marks, percentage, and display it
def calculate_total_and_percentage(marks):
    total_marks = sum(marks)
    percentage = total_marks / len(marks)
    return total_marks, percentage

# 3. Find the grade of the student
def find_grade(percentage):
    match percentage:
        case _ if percentage > 85:
            return "A"
        case _ if 75 <= percentage <= 85:
            return "B"
        case _ if 50 <= percentage < 75:
            return "C"
        case _ if 30 < percentage < 50:
            return "D"
        case _ if percentage <= 30:
            return "Reappear"
        case _:
            return "Invalid"

# Main program
def main():
    marks = get_student_marks()
    display_marks(marks)

    total_marks, percentage = calculate_total_and_percentage(marks)
    print(f"Total Marks: {total_marks}")
    print(f"Percentage: {percentage:.2f}%")

    grade = find_grade(percentage)
    print(f"Grade: {grade}")

if __name__ == "__main__":
    main()

```

```python
'''
1.14.	Consider the gravitational interactions between the Earth, Moon,and Sun in our solar system.
Given:
mass_earth =5.972e24 # Mass of Earth in kilograms
mass_moon =7.34767309e22 # Mass of Moon in kilograms
mass_sun =1.989e30 # Mass of Sun in kilograms

distance_earth_sun =1.496e11# Average distance between Earth and Sun in meters
distance_moon_earth = 3.844e8 #  Average distance between Moon and Earth in meters

Tasks:

Calculate the gravitational force between the Earth and the Sun.
Calculate the gravitational force between the Moon and the Earth.
Compare the calculated forces to determine which gravitational force is stronger.
Explain which celestial body (Earth or Moon) is more attracted to the other based on the comparison. '''


# Define the gravitational constant
G = 6.67430e-11  # in N * (m^2 / kg^2)

# Given data
mass_earth = 5.972e24  # in kg
mass_moon = 7.34767309e22  # in kg
mass_sun = 1.989e30  # in kg

distance_earth_sun = 1.496e11  # in m
distance_moon_earth = 3.844e8  # in m

# Calculate the gravitational force between Earth and Sun
force_earth_sun = G * (mass_earth * mass_sun) / (distance_earth_sun ** 2)

# Calculate the gravitational force between Moon and Earth
force_moon_earth = G * (mass_moon * mass_earth) / (distance_moon_earth ** 2)

# Determine which gravitational force is stronger
if force_earth_sun > force_moon_earth:
    stronger_force = "Earth and Sun"
else:
    stronger_force = "Moon and Earth"

# Display the results
print(f"Gravitational force between Earth and Sun: {force_earth_sun:.2e} N")
print(f"Gravitational force between Moon and Earth: {force_moon_earth:.2e} N")
print(f"The stronger gravitational force is between: {stronger_force}")

# Explanation of attraction
if force_earth_sun > force_moon_earth:
    attraction_explanation = "The Earth is more attracted to the Sun."
else:
    attraction_explanation = "The Moon is more attracted to the Earth."

print(attraction_explanation)
```

```python
'''
4. Create a simple banking system using object-oriented concepts in Python. Design classes representing
different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw,
and balance inquiry. Utlize inheritance to manage different account types efficiently.

Tasks:
1. Define base classes for bank accounts with common attributes and methods.
2. Implement subclasses for specific account types(eg. SavingsAccount,CheckingAccount)
3. Provide methods for deposit,withdraw,and balance inquiry in each subclass.
4. Test the banking system by creating instances of different account types and performing transactions. '''


class bankAccount:
  def __init__(self,account_holder_name,account_number,balance):
    self.__account_holder_name = account_holder_name
    self.__account_number = account_number
    self.__balance = balance

  def accountDetails(self):
    print("Account holder name:",self.__account_holder_name)
    print("Account number:",self.__account_number)

  def checkBalance(self):
    print("Account Balance:",self.__balance)

  def updateAccountHolderName(self,name):
    self.__account_holder_name = name

  def deposit(self,amount):
    if isinstance(amount,float) or isinstance(amount,int):
      if amount>0:
        self.__balance += amount
        print(f"{amount} deposited successfully")

      else:
       print('Error in depositing')
    else:
       print('Error in depositing')

  def withdraw(self,amount):
    if isinstance(amount,float) or isinstance(amount,int):
      if amount>0:
        if self.__balance >= amount :
          self.__balance -= amount
          print(f"{amount} withdrawn successfully")

        else:
          print(f"Balance is insufficient to withdraw {amount}")
          return
      else:
        print('Error in withdrawing')
    else:
       print('Error in withdrawing')

  def getBalance(self):
    return self.__balance

  def setBalance(self,amount):
    self.__balance = amount




class savingsAccount(bankAccount):
  def __init__(self,account_holder_name,account_number,balance):
    super().__init__(account_holder_name,account_number,balance)

  def balanceInquiry(self):
    super().checkBalance()
    return

  def accountDetails(self):
    super().accountDetails()
    return

  def deposit(self,amount):
    super().deposit(amount)

  def withdraw(self,amount):
    super().withdraw(amount)




class checkingAccount(bankAccount):
  def __init__(self,account_holder_name,account_number,balance):
    super().__init__(account_holder_name,account_number,balance)
    self.overdraft_limit = 1000

  def balanceInquiry(self):
    super().checkBalance()
    return

  def accountDetails(self):
    super().accountDetails()
    return

  def deposit(self,amount):
    super().deposit(amount)

  def withdraw(self,amount):
    if isinstance(amount,float) or isinstance(amount,int):
      if amount>0:
        if super().getBalance() + self.overdraft_limit >= amount :
          newBalance = super().getBalance()
          newBalance -=amount
          super().setBalance(newBalance)
          if super().getBalance() < 0:
            print(f'Overdraft! Balance due: Rs{super().getBalance()}')
            return
        else:
          print(f"Overdraft limit exceeded to withdraw {amount}")
          return
      else:
        print('Error in withdrawing')
    else:
       print('Error in withdrawing')
    print(f"{amount} withdrawn successfully")




a = savingsAccount('Rithin',10201,500)
a.accountDetails()
print()
a.balanceInquiry()
print()
a.deposit(200)
a.balanceInquiry()
print()
a.withdraw(10000)
print()
a.withdraw(20)
a.balanceInquiry()

print()
b = checkingAccount('Jithin',10200,1000)

b.accountDetails()
print()
b.balanceInquiry()
print()
b.deposit(200)
b.balanceInquiry()
print()
b.withdraw(10000)
print()
b.withdraw(20)
b.balanceInquiry()
print()
b.withdraw(1200)
b.balanceInquiry()
```

```python
'''
7. Write a code for Hotel Management System using OOPS:

Create a Room class that has attributes such as room number, room type, rate, and availability(private).

Implement methods to book a room, check in a guest, and check out a guest.

Use encapsulation to hide the room's unique identification number.

Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, each with their own
specific attributes and methods.'''

class Hotel:
    def __init__(self):
        self.rooms = []

    def display_rooms(self):
        for room in self.rooms:
            room.display_room()

    def add_room(self, room):
        self.rooms.append(room)

    def book_room(self, room_number):
        for room in self.rooms:
            if room.room_number == room_number:
                room.book_room()
                return

    def check_in(self, room_number):
        for room in self.rooms:
            if room.room_number == room_number:
                room.check_in()
                return

    def check_out(self, room_number):
        for room in self.rooms:
            if room.room_number == room_number:
                room.check_out()
                return


class Room:
    def __init__(self, room_number, room_type, rate, room_id, availability=True):
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__id = room_id
        self.__availability = availability

    def display_room(self):
        print('Room no:', self.room_number)
        print('Room type:', self.room_type)
        print('Rate:', self.rate)
        print(('Available' if self.__availability else 'Not available'))
        print()

    def book_room(self):
        if self.__availability:
            print(f"Room no:{self.room_number} booked successfully")
            self.__availability = False
        else:
            print(f"Room no:{self.room_number} isn't available")

    def check_in(self):
        if not self.__availability:
            print(f"Room no:{self.room_number} checked in")
        else:
            print(f"Room no:{self.room_number} is already available")

    def check_out(self):
        if not self.__availability:
            print(f"Room no:{self.room_number} checked out")
            self.__availability = True
        else:
            print(f"Room no:{self.room_number} is already available")


class SuiteRoom(Room):
    def __init__(self, room_number, room_type, rate, room_id, availability=True):
        super().__init__(room_number, room_type, rate, room_id, availability)

    def show_room(self):
        super().display_room()


class StandardRoom(Room):
    def __init__(self, room_number, room_type, rate, room_id, availability=True):
        super().__init__(room_number, room_type, rate, room_id, availability)

    def show_room(self):
        super().display_room()


# Example usage
h = Hotel()
r1 = SuiteRoom(room_number=1, room_type='Suite', rate=500, room_id=121)
r2 = StandardRoom(room_number=2, room_type='Standard', rate=1000, room_id=122)

h.add_room(r1)
h.add_room(r2)

h.book_room(room_number=1)
h.check_in(room_number=1)
h.check_out(room_number=1)
h.check_out(room_number=1)
print()

r2.display_room()
h.book_room(room_number=2)
h.book_room(room_number=2)

```

```python
'''
10. Write a code for Airline Reservation System using OOPS:

Create a Flight class that has attributes such as flight number, departure,and arrival airports,departure and
arrival times, and available seats(private).

Implement methods to book a seat,cancel a reservation,and get the remaining available seats.

Use encapsulation to hide the flight's unique indentification number.

Inherit from the Flight class to create a DomesticFlight class and an InternationalFlight class, each with their
own specific attributes and methods.'''



class AirlineReservationSystem:
    def __init__(self):
        self.flights = []

    def add_flight(self, flight):
        if flight not in self.flights:
            self.flights.append(flight)
            flight.add_flight()
        else:
            print(f"Flight {flight.flight_number} already exists")

    def book_seat(self, flight_id):
        for flight in self.flights:
            if flight.id() == flight_id:
                flight.book_seat()
                return

    def cancel_reservation(self, flight_id):
        for flight in self.flights:
            if flight.id() == flight_id:
                flight.cancel_reservation()
                return

    def remaining_seats(self, flight_id):
        for flight in self.flights:
            if flight.id() == flight_id:
                flight.remaining_seats()
                return


class Flight:
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, flight_id):
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seats = available_seats
        self.__id = flight_id

    def id(self):
        return self.__id

    def show_flight(self):
        print("Flight number:", self.flight_number)
        print("Departure airport:", self.departure_airport)
        print("Arrival airport:", self.arrival_airport)
        print("Departure time:", self.departure_time)
        print("Arrival time:", self.arrival_time)
        print("Available seats:", self.__available_seats)

    def add_flight(self):
        print(f"Flight {self.flight_number} added")

    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked on flight {self.flight_number}. Seats left: {self.__available_seats}")
        else:
            print(f"No seats available on flight {self.flight_number}")

    def cancel_reservation(self):
        self.__available_seats += 1
        print(f"Reservation cancelled on flight {self.flight_number}. Seats available: {self.__available_seats}")

    def remaining_seats(self):
        print(f"Remaining seats on flight {self.flight_number}: {self.__available_seats}")
        return


class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, flight_id, domestic_code):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, flight_id)
        self.domestic_code = domestic_code

    def show_flight(self):
        super().show_flight()
        print("Domestic code:", self.domestic_code)


class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, flight_id, international_code):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, flight_id)
        self.international_code = international_code

    def show_flight(self):
        super().show_flight()
        print("International code:", self.international_code)


# Example usage
ars = AirlineReservationSystem()
df1 = DomesticFlight(flight_number="AI101", departure_airport="JFK", arrival_airport="LAX", departure_time="2024-08-01 08:00", arrival_time="2024-08-01 11:00", available_seats=10, flight_id=1, domestic_code="D001")
if1 = InternationalFlight(flight_number="AI201", departure_airport="JFK", arrival_airport="LHR", departure_time="2024-08-01 08:00", arrival_time="2024-08-01 20:00", available_seats=20, flight_id=2, international_code="I001")

ars.add_flight(df1)
ars.add_flight(if1)

df1.show_flight()
if1.show_flight()

ars.book_seat(flight_id=1)
ars.book_seat(flight_id=1)

ars.remaining_seats(flight_id=1)

ars.cancel_reservation(flight_id=1)

ars.remaining_seats(flight_id=1)

ars.book_seat(flight_id=2)
ars.remaining_seats(flight_id=2)
```

```python
'''
18. Create a python script that reads a text file named "expenses.txt" and calculates the total amount spent on various expenses listed in the file.'''

filepath = "expenses.txt"
with open(filepath,'w')as f:
  f.write("Rent:1200")
with open(filepath,'a')as f:
  f.write('\nGroceries:150')
  f.write('\nUtilities:100')
  f.write('\nTransportation:80')
  f.write('\nEntertainment:50')


def find_total(filepath):
  with open(filepath,'r') as f:
    expenses = f.read()
    expenses = expenses.split('\n')
    total_amount = 0

    for expense in expenses:
      expense = float(expense.split(":")[-1])
      total_amount+=expense

    return total_amount

print("Total expenses:",find_total(filepath))
```

```python
'''
19. Create a python program that reads a text file named "paragraph.txt" and counts the occurences of
each word in the paragraph, displaying the results in alphabetical order.'''

filepath = "paragraph.txt"
with open(filepath,'w')as f:
  f.write("This is a sample paragraph. This paragraph is for testing purposes.")

from collections import Counter
import re

def count_words(file_path):
    word_count = Counter()

    try:
        with open(file_path, 'r') as file:
            text = file.read().lower()  # Read and convert to lower case
            # Remove punctuation using regex and split into words
            words = re.findall(r'\b\w+\b', text)
            word_count.update(words)
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
    except IOError as e:
        print(f"Error reading file: {e}")

    return dict(sorted(word_count.items()))

def display_word_count(word_count):

    for word, count in sorted(word_count.items()):
        print(f"{word}: {count}")

# Example usage
if __name__ == "__main__":
    file_path = 'paragraph.txt'
    word_count = count_words(file_path)
    display_word_count(word_count)
```
